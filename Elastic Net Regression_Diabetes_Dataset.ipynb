```python
X = data.drop(columns=["Outcome"])
```

```python
y = data.Outcome
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Ridge,Lasso
from sklearn.metrics import mean_absolute_error, mean_squared_error,r2_score
from sklearn.linear_model import ElasticNet
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y, 
                                                    test_size=0.2,
                                                    random_state=75)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
#Ridge Regression
ridge_model= Ridge(alpha=0.5)
```

```python
ridge_model.fit(X_train, y_train)
```

```python
#predict Test Data
y_pred = ridge_model.predict(X_test)
```

```python
ridge_model.score(X_test,y_test)*100
```

```python
r2_score(y_test,y_pred)
```

```python
#Ridge Regression
lasso_model= Lasso(alpha=0.5)
```

```python
lasso_model.fit(X_train, y_train)
```

```python
#predict Test Data
y_pred = lasso_model.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
#Evaluating the Model
MAE= mean_absolute_error(y_test, y_pred)
MSE= mean_squared_error(y_test, y_pred)
RMSE= np.sqrt(MSE)
```

```python
pd.DataFrame([MAE, MSE, RMSE], index=['MAE', 'MSE', 'RMSE'], columns=['metrics'])
```

```python
ridge_model.coef_
```

```python
feature_names = X.columns
coefficients = lasso_model.coef_
```

```python
# Create a Series for easier analysis
coeff_df = pd.Series(coefficients, index=feature_names).sort_values(ascending=False)
```

```python
# Print all coefficients
print(coeff_df)
```

```python
count = 0
for i in coeff_df:
    if i!=0:
        count+=1
print("After Lasso, we have only ", count , " none-zero coefficients.")
```

```python
# Create an ElasticNet model instance with specific hyperparameters
# l1_ratio=0.5 means an equal mix of L1 and L2 penalties
elasticnet_model = ElasticNet(alpha=0.1, l1_ratio=0.5, random_state=42)
```

```python
elasticnet_model.fit(X_train, y_train)
```

```python
#predict Test Data
y_pred = elasticnet_model.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
list_elastic_model_coef=[]
list_elastic_model_coef = elasticnet_model.coef_
elasticnet_model.coef_
```

```python
count = 0
for i in list_elastic_model_coef:
    if i!=0:
        count+=1
print("After Elastic Net, we have only ", count , " none-zero coefficients.")
```
