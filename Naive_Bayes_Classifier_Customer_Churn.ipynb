```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore') # handle_unknown='ignore' for unseen categories in test data
```

```python
categorical_columns = ['gender', 'Partner','Dependents','PhoneService','StreamingMovies','PaperlessBilling']
```

```python
encoded_features = encoder.fit_transform(data[categorical_columns])
```

```python
data_encoded = pd.DataFrame(encoded_features, columns=encoder.get_feature_names_out(categorical_columns))
```

```python
data_processed = data.drop(columns=categorical_columns) #removing all original columns
```

```python
data_final = pd.concat([data_processed, data_encoded], axis=1)
```

```python
data_final.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder()
data_final[["MultipleLines","InternetService","OnlineSecurity","OnlineBackup","DeviceProtection","TechSupport","StreamingTV","Contract","PaymentMethod","Churn"]]=oe.fit_transform(data_final[["MultipleLines","InternetService","OnlineSecurity","OnlineBackup","DeviceProtection","TechSupport","StreamingTV","Contract","PaymentMethod","Churn"]])
```

```python
data_final.head()
```

```python
data_final.drop(columns="customerID",axis=1,inplace=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
#Split data
X = data_final.drop("Churn", axis=1)
y = data_final["Churn"]

X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.3,
                                                    random_state=42)
```

```python
X.head()
```

```python
X = X.replace(r'^\s*$', np.nan, regex=True)
```

```python
X_train["TotalCharges"] = pd.to_numeric(X_train["TotalCharges"], errors="coerce")
```

```python
X_test["TotalCharges"] = pd.to_numeric(X_test["TotalCharges"], errors="coerce")
```

```python
X_train["TotalCharges"] = X_train["TotalCharges"].fillna(X_train["TotalCharges"].mean())
```

```python
X_test["TotalCharges"] = X_test["TotalCharges"].fillna(X_test["TotalCharges"].mean())
```

```python
#Train Gaussian Naive Bayes
gnb = GaussianNB()
```

```python
gnb.fit(X_train, y_train)
```

```python
#Predictions
y_pred = gnb.predict(X_test)
print("Predictions:", y_pred)
```

```python
#Evaluation Metrics
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
