```python
#Import Libraries
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#We need to decide best columns for consideration
X = data.drop(columns=["mpg"])
y = data['mpg']
```

```python
#While standardising found dataset contains ?, finding it from datset on which column has that
X[X.isin(["?"]).any(axis=1)]
```

```python
#replacing with nan
X['horsepower'] = X['horsepower'].replace('?', np.nan)
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
```

```python
X[['horsepower']] = imputer.fit_transform(X[['horsepower']])
```

```python
#Scaling features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#train test split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42)
```

```python
X_train.shape
```

```python
#Build KNN Regressor
knn = KNeighborsRegressor(n_neighbors=5)
knn.fit(X_train, y_train)
```

```python
#Predict
y_pred = knn.predict(X_test)
```

```python
y_pred.shape
```

```python
print("MSE:", mean_squared_error(y_test, y_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
print("R2 Score:", r2_score(y_test, y_pred))
```

```python
#k-fold validation
scores = cross_val_score(knn, X_scaled, y, cv=5, scoring='r2')
print("CV R2 Scores:", scores)
print("Mean R2:", scores.mean())
```

```python
#Hyperparameter Tuning using GridSearchCV
param_grid = {
    'n_neighbors': list(range(1, 21)),
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan']
}

grid = GridSearchCV(
    KNeighborsRegressor(),
    param_grid,
    cv=5,
    scoring='neg_mean_squared_error'
)

grid.fit(X_scaled, y)

print("Best Params:", grid.best_params_)
print("Best Score (MSE):", -grid.best_score_)
```

```python
#Actual vs Predicted
plt.figure(figsize=(10,6))
plt.plot(y_test.values, label='Actual', linewidth=3)
plt.plot(y_pred, label='Predicted', linewidth=3)

plt.xlabel("Samples")
plt.ylabel("Target Value")
plt.title("Actual vs Predicted Trend")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#predicted with residuals/errors
residuals = y_test - y_pred

plt.figure(figsize=(8,6))
plt.scatter(y_pred, residuals, c=residuals, cmap='coolwarm', s=80)
plt.axhline(0, color='black', linestyle='--')

plt.xlabel("Predicted")
plt.ylabel("Residuals")
plt.title("Residual Plot")
plt.colorbar(label="Residual Size")
plt.grid(True)
plt.show()
```

```python
#Error distribution
errors = y_test - y_pred
sns.histplot(errors, kde=True)
plt.title("Error Distribution")
plt.show()
```
