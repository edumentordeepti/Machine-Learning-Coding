```python
data.drop(columns=['ins_premium', 'ins_losses', 'abbrev'],inplace=True)# RUN ONLY ONCE 
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(data.drop(columns="total"),
                           data.total,
                          random_state=42,
                          test_size=0.1)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train) # Training on 90% data x,y
```

```python
y_pred=lr.predict(x_test) # prdict 10% of independent data
```

```python
y_pred
```

```python
y_test - y_pred
```

```python
lr.score(x_test,y_test)*100
```

```python
from sklearn.metrics import r2_score #R-Squared
```

```python
r2_score(y_test,y_pred)
```

```python
from sklearn.metrics import mean_absolute_error #Mean Absolute Error
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
from sklearn.metrics import mean_squared_error #Mean Square Error
```

```python
mean_squared_error(y_test,y_pred)
```

```python
rmse = np.sqrt(mean_squared_error(y_test,y_pred))
print(rmse)
```
